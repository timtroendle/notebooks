# Which impact does heat gain from people have on energy usage in buildings?

To analyse the impact of heat gains from people I ran the same building simulation twice, once with and once without heat gain. The heat gain was a constant 140W all year long. 

Eplus version 3.1

Weather file: Chicago OHare provided by Eplus

```python
from pathlib import Path

import pandas as pd
%matplotlib inline
```

```python
path_to_results_including_heat_gain = Path('/Users/Trondle/Sync/cambridge/eplus-test-run/G9998738Rt-metabolismMeter.csv')
path_to_results_without_heat_gain = Path('/Users/Trondle/Sync/cambridge/eplus-test-run/G9998738Rt-no-metabolismMeter.csv')
```

```python
heat_gain = pd.read_csv(path_to_results_including_heat_gain, index_col=0)
no_heat_gain = pd.read_csv(path_to_results_without_heat_gain, index_col=0)
```

```python
heat_gain
```

```python
heat_gain.plot(figsize=(14, 7))
```

```python
no_heat_gain.plot(figsize=(14, 7))
```

```python
(heat_gain - no_heat_gain).plot(figsize=(14, 7))
```

```python
no_heat_gain.sum()/heat_gain.sum()
```

It seems the impact of heat gain from people on heating is rather low, less than 2% increase in this case. The impact on cooling seems to be a bit bigger, more than 6% reduction in this case.

Obviously this is not representative and only an indication.
